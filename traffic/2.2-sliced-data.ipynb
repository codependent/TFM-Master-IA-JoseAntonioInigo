```python
# Configuration
MONTHS = 14
```

```python
import pandas as pd
from dateutil.relativedelta import relativedelta

df = pd.read_csv("dataset/processed/traffic-meteo.csv", parse_dates=["fecha"])

# Compute the cutoff date: MONTHS before the global end date
end_date = df["fecha"].max()
cutoff_date = end_date - relativedelta(months=MONTHS)

print(f"End date:    {end_date}")
print(f"Cutoff date: {cutoff_date} ({MONTHS} months back)")
print(f"Original rows: {len(df):,}")

# Keep only rows strictly after the cutoff per sensor
df_sliced = df[df["fecha"] > cutoff_date].copy()

print(df_sliced.head())

print(f"Sliced rows:   {len(df_sliced):,}")
print(f"Sensors:       {df_sliced['id'].nunique()}")

# Save
output_path = f"dataset/processed/traffic-meteo-{MONTHS}-slice.csv"
df_sliced.to_csv(output_path, index=False)
print(f"Saved to {output_path}")
```
